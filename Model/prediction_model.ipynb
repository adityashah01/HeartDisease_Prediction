```python
import pandas as pd
import joblib
from tensorflow.keras.models import load_model

# Folder containing the saved project outputs
MODEL_FOLDER = "/content/drive/MyDrive/Heart_Disease_Project/outputs"

# Load trained ANN
loaded_model = load_model(
    f"{MODEL_FOLDER}/tuned_ann.keras"
)

# Load the same imputer and scaler used during training
loaded_preprocessor = joblib.load(
    f"{MODEL_FOLDER}/ann_preprocessor.joblib"
)

print("ANN model and preprocessor loaded successfully!")
```

```python
import pandas as pd

print("=" * 65)
print("      HEART DISEASE RISK CLASSIFICATION — ANN DEMO")
print("=" * 65)

print("\nEnter 0 for No and 1 for Yes unless stated otherwise.\n")

# All 21 features must be entered
patient_data = {
    "HighBP": int(input("High blood pressure? (0=No, 1=Yes): ")),

    "HighChol": int(input("High cholesterol? (0=No, 1=Yes): ")),

    "CholCheck": int(input("Cholesterol checked within 5 years? (0=No, 1=Yes): ")),

    "BMI": float(input("BMI value, for example 25: ")),

    "Smoker": int(input("Smoked at least 100 cigarettes previously? (0=No, 1=Yes): ")),

    "Stroke": int(input("History of stroke? (0=No, 1=Yes): ")),

    "Diabetes": int(input(
        "Diabetes category (0=No, 1=Prediabetes, 2=Diabetes): "
    )),

    "PhysActivity": int(input(
        "Physical activity during the past 30 days? (0=No, 1=Yes): "
    )),

    "Fruits": int(input("Consumes fruit regularly? (0=No, 1=Yes): ")),

    "Veggies": int(input("Consumes vegetables regularly? (0=No, 1=Yes): ")),

    "HvyAlcoholConsump": int(input(
        "Heavy alcohol-consumption category? (0=No, 1=Yes): "
    )),

    "AnyHealthcare": int(input(
        "Has healthcare coverage? (0=No, 1=Yes): "
    )),

    "NoDocbcCost": int(input(
        "Unable to visit a doctor because of cost? (0=No, 1=Yes): "
    )),

    "GenHlth": int(input(
        "General health (1=Excellent, 2=Very good, 3=Good, "
        "4=Fair, 5=Poor): "
    )),

    "MentHlth": int(input(
        "Days of poor mental health during past 30 days (0-30): "
    )),

    "PhysHlth": int(input(
        "Days of poor physical health during past 30 days (0-30): "
    )),

    "DiffWalk": int(input(
        "Difficulty walking or climbing stairs? (0=No, 1=Yes): "
    )),

    "Sex": int(input("Sex category (0=Female, 1=Male): ")),

    "Age": int(input(
        "Age category (1=18-24, 2=25-29, 3=30-34, 4=35-39, "
        "5=40-44, 6=45-49, 7=50-54, 8=55-59, 9=60-64, "
        "10=65-69, 11=70-74, 12=75-79, 13=80+): "
    )),

    "Education": int(input(
        "Education category (1-6, where 6=College graduate): "
    )),

    "Income": int(input(
        "Income category (1-8, from lowest to highest): "
    ))
}

# Required training feature order
FEATURES = [
    "HighBP", "HighChol", "CholCheck", "BMI", "Smoker", "Stroke",
    "Diabetes", "PhysActivity", "Fruits", "Veggies",
    "HvyAlcoholConsump", "AnyHealthcare", "NoDocbcCost",
    "GenHlth", "MentHlth", "PhysHlth", "DiffWalk", "Sex",
    "Age", "Education", "Income"
]

# Convert entered information into a one-row DataFrame
patient_df = pd.DataFrame(
    [patient_data],
    columns=FEATURES
)

# Apply the same preprocessing used during training
patient_scaled = loaded_preprocessor.transform(
    patient_df
).astype("float32")

# Obtain ANN probability
probability = float(
    loaded_model.predict(
        patient_scaled,
        verbose=0
    )[0, 0]
)

# Convert probability into class using threshold 0.50
predicted_class = int(probability >= 0.50)

print("\n" + "=" * 65)
print("                    PREDICTION RESULT")
print("=" * 65)

print(f"Positive-class probability: {probability * 100:.2f}%")
print(f"Predicted class: {predicted_class}")

if predicted_class == 1:
    print("Model classification: POSITIVE RISK CLASS")
else:
    print("Model classification: NEGATIVE RISK CLASS")

print("\nImportant: This is an academic ML demonstration.")
print("It is not a medical diagnosis.")
```
